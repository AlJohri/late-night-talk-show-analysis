# Colbert vs Fallon

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

ratings = pd.read_csv('nielson/ratings.csv', parse_dates=['start_date', 'end_date'])
```

```python
ratings.show.unique()
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ratings[ratings.show == "Late Show with Stephen Colbert"].plot(x='start_date', y='viewers_week', ax=ax, marker='o', label='colbert')
ratings[ratings.show == "The Tonight Show Starring Jimmy Fallon"].plot(x='start_date', y='viewers_week', ax=ax, marker='o', label='fallon')
```
