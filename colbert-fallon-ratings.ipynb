# Colbert vs Fallon Rating Analysis

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

nielson = pd.read_csv('nielson/ratings.csv', parse_dates=['start_date', 'end_date'])
colbert_trump_mentions = pd.read_csv('the-late-show-with-stephen-colbert/data/trump_mentions.csv',
                                     parse_dates=['air_date'])
fallon_trump_mentions = pd.read_csv('the-tonight-show-starting-jimmy-fallon/data/trump_mentions.csv',
                                     parse_dates=['air_date'])
```

```python
nielson.show.unique()
```

## Number of Viewers per Week in Millions

```python
fig, ax = plt.subplots(figsize=(15,6))
nielson[nielson.show == "Late Show with Stephen Colbert"].set_index('start_date').plot(y='viewers_week', ax=ax, marker='o', label='colbert')
nielson[nielson.show == "The Tonight Show Starring Jimmy Fallon"].set_index('start_date').plot(y='viewers_week', ax=ax, marker='o', label='fallon')
```

## Number of Trump Mentions per Episode average by Month

```python
fig, ax = plt.subplots(figsize=(15,6))
colbert_trump_mentions.set_index('air_date').resample('M').mean().plot(y='count', ax=ax, marker='o', label='colbert')
fallon_trump_mentions.set_index('air_date').resample('M').mean().plot(y='count', ax=ax, marker='o', label='fallon')
```
